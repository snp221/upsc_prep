```python
import requests
from bs4 import BeautifulSoup
import time

url_base = 'https://www.drishtiias.com/model-essays'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

all_links = []
page = 0

while True:
    # Construct URL
    url = f"{url_base}?page={page}" if page > 0 else url_base
    print(f"Scraping page {page}: {url}")
    
    response = requests.get(url, headers=headers)
    
    if response.status_code != 200:
        print(f"Failed to fetch page {page}")
        break
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find div with class 'list-category'
    list_divs = soup.find_all('div', class_='list-category')
    
    if not list_divs:
        print(f"No more content found on page {page}")
        break
    
    # Extract links from ol tags within each div
    for div in list_divs:
        ol_tags = div.find_all('ol')
        for ol in ol_tags:
            links = ol.find_all('a', href=True)
            for link in links:
                href = link['href']
                text = link.get_text(strip=True)
                
                # Make URL absolute if it's relative
                if href.startswith('/'):
                    href = f"https://www.drishtiias.com{href}"
                
                all_links.append({
                    'text': text,
                    'url': href
                })
                print(f"  Found: {text}")
    
    page += 1
    time.sleep(1)  # Be polite to server

print(f"\n✅ Total links collected: {len(all_links)}")

# Display all links
for item in all_links:
    print(f"{item['text']}: {item['url']}")
```

```python
all_details = []
import requests
from bs4 import BeautifulSoup
import re
for item in all_links:    
    

    # Example URL - replace with actual essay URL
    url = item['url']
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}

    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Get title
    title_tag = soup.find('h1', id='dynamic-title')
    title = title_tag.get_text(strip=True) if title_tag else 'No title found'

    # Get content using CSS selector
    content_div = soup.select_one('body > section > div.wrapper > div > article > div:nth-child(2) > div.article-detail > div.wrapper > div')

    if content_div:
        # Remove unwanted elements
        for element in content_div.find_all(['script', 'style', 'img', 'table']):
            element.decompose()
        
        # Extract text
        content_text = content_div.get_text(separator=' ', strip=True)
        content_text = re.sub(r'\s+', ' ', content_text)
    else:
        content_text = 'No content found'

    all_details.append({
        'title': title, 
        'content': content_text
    })

    print(f"Title: {title}\n")
```

```python
import json
with open('essay_titles.json', 'w', encoding='utf-8') as f:
    json.dump(all_details, f, ensure_ascii=False, indent=2)
```
