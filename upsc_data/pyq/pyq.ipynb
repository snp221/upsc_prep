```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
url = 'https://pwonlyias.com/mains-solved-papers-by-subject/'
questions = set()
# Send a GET request to the webpage
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
divs = soup.find_all('div', class_='container')
for link in soup.find_all('a', href=True):
    href = link['href']
    if '/mains-solved-papers-by-subject/' in href:
        print(href)
        questions.add(href)
```

```python
for question in questions:
    response = requests.get(question, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    links = []
    for a_tag in soup.find_all('a', href=True):
        span = a_tag.find('span')
        if span and span.text.strip().startswith('Q.'):
            links.append({
                'url': a_tag['href'],
                'question_number': span.text.strip(),
                'title': a_tag.find('h4').text.strip() if a_tag.find('h4') else ''
            })

for link in links:
    print(link)
```

```python
import re
all_questions_data = []
for question_url in links:
    question_url = question_url['url']
    response = requests.get(question_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    question_text = None
    question_h4 = soup.find('h4')
    if question_h4:
        # Remove "Que. " or "Q. " prefix if exists
        question_text = question_h4.get_text(strip=True)
        question_text = re.sub(r'^(Que\.|Q\.)\s*', '', question_text)


year = None
year_span = soup.find('span', string=lambda text: text and 'UPSC CSE :' in text)
if year_span:
    year_text = year_span.get_text(strip=True)
    # Extract just the year number
    year = year_text.split(':')[-1].strip()

content_div = soup.find('div', class_='pf-content')
        
if content_div:
    # Remove unwanted elements
    for unwanted in content_div.find_all(['div', 'a', 'img']):
        unwanted.decompose()
    
    # Get text with space separator
    content_text = content_div.get_text(separator=' ', strip=True)
    
    # Clean up excessive whitespace
    content_text = re.sub(r'\s+', ' ', content_text)
    
    # Optional: Replace multiple spaces/newlines with single newline
    content_text = re.sub(r'\s{2,}', '\n', content_text)
    
    all_questions_data.append({
        'year': year,
        'question': question_text,
        'content': content_text
    })

print(all_questions_data)
```

```python
import requests
from bs4 import BeautifulSoup
import json
import os
import re
import time

# Configuration
base_path = r'C:\Users\sv221\OneDrive\Desktop\django_tutorial\upsc_prep\upsc_data\pyq'
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
url = 'https://pwonlyias.com/mains-solved-papers-by-subject/'

# Step 1: Get all topic URLs
print("Fetching topic URLs...")
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

topic_urls = set()
for link in soup.find_all('a', href=True):
    href = link['href']
    if '/mains-solved-papers-by-subject/' in href and href != url:
        topic_urls.add(href)
        print(f"  Found: {href}")

print(f"\nTotal topics: {len(topic_urls)}\n")

# Step 2: Process each topic
for topic_url in topic_urls:
    # Extract topic name from URL
    topic_name = topic_url.rstrip('/').split('/')[-1]
    
    print(f"{'='*80}")
    print(f"üìÅ Processing: {topic_name}")
    print(f"{'='*80}")
    
    # Create topic folder
    topic_folder = os.path.join(base_path, topic_name)
    os.makedirs(topic_folder, exist_ok=True)
    
    # Get all question links for this topic
    response = requests.get(topic_url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    question_links = []
    for a_tag in soup.find_all('a', href=True):
        span = a_tag.find('span')
        if span and span.text.strip().startswith('Q.'):
            question_links.append({
                'url': a_tag['href'],
                'question_number': span.text.strip(),
                'title': a_tag.find('h4').text.strip() if a_tag.find('h4') else ''
            })
    
    print(f"Found {len(question_links)} questions")
    
    # Step 3: Extract content from each question
    all_questions_data = []
    
    for idx, link_data in enumerate(question_links, 1):
        question_url = link_data['url']
        
        print(f"  [{idx}/{len(question_links)}] {link_data['question_number']}: {link_data['title'][:50]}...")
        
        try:
            response = requests.get(question_url, headers=headers)
            soup = BeautifulSoup(response.content, 'html.parser')
            
            # Extract question
            question_text = None
            question_h4 = soup.find('h4')
            if question_h4:
                question_text = question_h4.get_text(strip=True)
                question_text = re.sub(r'^(Que\.|Q\.)\s*', '', question_text)
            
            # Extract year
            year = None
            year_span = soup.find('span', string=lambda text: text and 'UPSC CSE :' in text)
            if year_span:
                year_text = year_span.get_text(strip=True)
                year = year_text.split(':')[-1].strip()
            
            # Extract content
            content_div = soup.find('div', class_='pf-content')
            
            if content_div:
                # Remove unwanted elements
                for unwanted in content_div.find_all(['div', 'a', 'img']):
                    unwanted.decompose()
                
                # Get text with space separator
                content_text = content_div.get_text(separator=' ', strip=True)
                
                # Clean up excessive whitespace
                content_text = re.sub(r'\s+', ' ', content_text)
                
                all_questions_data.append({
                    'question': question_text,
                    'year': year,
                    'topic': topic_name,
                    'answer': content_text
                })
                
                print(f"      ‚úì Year: {year}")
            else:
                print(f"      ‚úó No content found")
            
            # Be polite to the server
            time.sleep(1)
            
        except Exception as e:
            print(f"      ‚úó Error: {str(e)}")
            continue
    
    # Step 4: Save to JSON file
    if all_questions_data:
        json_file = os.path.join(topic_folder, f'{topic_name}.json')
        with open(json_file, 'w', encoding='utf-8') as f:
            json.dump(all_questions_data, f, ensure_ascii=False, indent=2)
        
        print(f"\nüíæ Saved {len(all_questions_data)} questions to {topic_name}.json")
    else:
        print(f"\n‚ö†Ô∏è No questions extracted for {topic_name}")
    
    print()

print(f"\n{'='*80}")
print("‚úÖ All topics processed!")
print(f"{'='*80}")
```
