```python
import requests
from bs4 import BeautifulSoup

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
url = 'https://pwonlyias.com/ncert-notes/'  # Replace with actual URL

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')

# Find the subjects row
subjects_row = soup.find('div', class_='subjects-row')

subject_links = []

if subjects_row:
    # Find all links within the row
    for a_tag in subjects_row.find_all('a', href=True):
        h3 = a_tag.find('h3')
        if h3:
            subject_links.append({
                'subject': h3.get_text(strip=True),
                'url': a_tag['href']
            })

for subject in subject_links:
    print(subject)
```

```python
import re
all_details = []
for topic in all_topics:
    url = topic['url']  # or the actual URL
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')

    content_div = soup.find('div', class_='pf-content')

    if content_div:
        # Remove printfriendly div
        for div in content_div.find_all('div', class_='printfriendly'):
            div.decompose()
        
        # Remove pf-button divs
        for div in content_div.find_all('div', class_=['pf-button', 'vc_gray_box']):
            div.decompose()
        
        # Remove images
        for img in content_div.find_all('img'):
            img.decompose()
        
        # Remove scripts
        for script in content_div.find_all('script'):
            script.decompose()
        
        # Get clean text
        content_text = content_div.get_text(separator=' ', strip=True)
        
        # Clean whitespace
        content_text = re.sub(r'\s+', ' ', content_text)
        
        all_details.append({
                    'subject': topic['subject'],
                    'category': topic['category'],
                    'content': content_text
                })
        print(f"✅ Fetched content for: {topic['url']}")
        
print(f"\n✅ Total topics: {len(all_details)}")

# Save to JSON
with open('ncert_topics.json', 'w', encoding='utf-8') as f:
    json.dump(all_details, f, ensure_ascii=False, indent=2)
```
